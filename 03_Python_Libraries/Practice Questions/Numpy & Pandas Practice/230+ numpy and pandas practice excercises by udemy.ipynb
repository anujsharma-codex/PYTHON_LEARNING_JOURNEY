```python
import numpy as np
A = np.array([[3, 2, 1, 4],
              [5, 2, 1, 6]])
B = np.array([[3, 2, 1, 4],
              [5, 2, 0, 6]])
C = np.array([[True, False, False],
              [True, True, True]])
D = np.array([0.1, 0.3])

for name,array in zip(list('ABCD'),[A,B,C,D]):
    print(f"{name}:{np.all(array)}")
```

```python
import numpy as np

A = np.array([[3, 2, 1, 4],
              [5, 2, 1, 6]])
B = np.array([[3, 2, 1, 4],
              [5, 2, 0, 6]])
C = np.array([[True, False, False],
              [True, True, True]])

for name,value in zip(list('ABC'),[A,B,C]):
    print(f"{name}: {np.all(value,axis=1)}")
```

```python
A = np.array([[3, 2, 1, np.nan],
              [5, np.nan, 1, 6]])
print(np.isnan(A))
```

```python
A = np.array([0.4, 0.5, 0.3])
B = np.array([0.39999999, 0.5000001, 0.3])
print(np.allclose(A,B))
```

```python
A = np.array([0.4, 0.5, 0.3])
B = np.array([0.3999999999, 0.5000000001, 0.3])

comparision=(A==B)
print(comparision)

# other way
print(np.equal(A,B))
```

```python
A = np.array([0.4, 0.5, 0.3, 0.9])
B = np.array([0.38, 0.51, 0.3, 0.91])
# way 1
greater_A=np.where(A>B,True, False)
print(greater_A)

# way 2
print(A>B)

# way 3
print(np.greater(A,B))
```
